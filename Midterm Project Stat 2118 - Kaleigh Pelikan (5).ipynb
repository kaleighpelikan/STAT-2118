```r
plot(fertility ~ log(ppgdp), data = data)
abline(lm(fertility ~ log(ppgdp), data = data))

reg = lm(fertility ~ log(ppgdp), data = data)
summary(reg)
```

Inferences about the Slope B1 when alpha = 0.05

```r
#t* = -14.61

abs(-14.61)
```

```r
qt(0.975, 197)
```

Because t* = 14.61 > t(0.975, 197) = 1.97, we can reject the null hypothesis and conclude that the slope coefficient is significantly different from 0. This means that there is a significant correlation between the variables of log(ppgdp) and fertility.
