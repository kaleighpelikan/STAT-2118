# Question 3

```r
reg1 = lm(fertility ~ log(ppgdp), data = data)
summary(reg1)
```

```r
reg2 = lm(fertility ~ lifeExpF, data = data)
summary(reg2)
```

```r
reg3 = lm(fertility ~ pctUrban, data = data)
summary(reg3)
```

If we were to select only one variable to predict fertility, I would choose lifeExpF. This is because the independent variable lifeExpF has the highest adjusted R-squared value. The adjusted R-squared value tells us how much of the variability in the data can be attributed to the relationship between variables. The adjusted R-squared value for the relationship between fertility and lifeExpF is 0.6767. The is higher than the R-Squared value for fertility and ppgdp (0.5175) and the R-Squared value for fertility and pctUrban (0.2913). Because of this, if we were to select only one variable to predict fertility, I would choose lifeExpF.

If we were to use two variables to predict fertility, I would choose log(ppgdp) and lifeExpF. This is because as seen in mreg1, the multiple linear regression for the relationship between fertility and log(ppgdp) + lifeExpF has the highest adjusted R-squared value at 0.6931. This means that 69.31% of the variability in the data can be attributed to the relationship between fertility and log(ppgdp) + lifeExpF. This adjusted R-Squared value is higher than for pctUrban + lifeExpF (0.6786) and is also higher than for log(ppgdp) + pctUrban (0.5151). Because it has the highest R-Squared value, if we use two variables to predict fertility, I would select log(ppgdp) and lifeExpF.

# Question 6

```r
interaction1 = lm(fertility ~ lifeExpF + log(ppgdp) + pctUrban + lifeExpF*log(ppgdp), data = data)
summary(interaction1)

ols_regress(interaction1)
```

Because the p value for lifeExpF:ppgdp = 0.000 < alpha = 0.05, we should reject the null hypothesis. This means that there is interaction between lifeExpF and log(ppgdp).

```r
interaction2 = lm(fertility ~ lifeExpF + pctUrban + lifeExpF*pctUrban, data = data)
summary(interaction2)

ols_regress(interaction2)
```

Because the p value for lifeExpF:pctUrban = 0.001 < alpha = 0.05, we should reject the null hypothesis. This means that there is interaction between lifeExpF and pctUrban.

```r

interaction3 = lm(fertility ~ lifeExpF + pctUrban + log(ppgdp) + log(ppgdp)*pctUrban, data = data)
summary(interaction3)

ols_regress(interaction3)
```

Because the p value for pctUrban:ppgdp = 0.000 < alpha = 0.05, we should reject the null hypothesis. This means that there is interaction between pctUrban and log(ppgdp).

# Question 7

```r
qreg1 = lm(fertility ~ log(ppgdp) + I(log(ppgdp)^2), data  = data)
summary(qreg1)

ols_regress(qreg1)
```

Because the p-value = 0.000 < alpha = 0.05, we should reject the null. This means that there is a quadratic relationship.

```r
qreg2 = lm(fertility ~ pctUrban + I(pctUrban^2), data  = data)

summary(qreg2)

ols_regress(qreg2)
```

Because the p-value = 0.044 < alpha = 0.05, we should reject the null. This means that there is a quadratic relationship.

```r
qreg3 = lm(fertility ~ lifeExpF + I(lifeExpF^2), data  = data)
summary(qreg3)

ols_regress(qreg3)
```

Because the p-value = 0.510 > alpha = 0.05, we should fail to reject the null. This means that there is no quadratic relationship.
